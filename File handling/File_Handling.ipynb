# read operation => 'r'

- read data from text file

```python
f = open(r"/content/data", mode='r')
data = f.read()
print(type(data))
f.close()
```

# write operation => 'w'
- If file exist then data will be overwrite.
- if not then new file will be created and data will be written in that file.

```python
f = open("/content/data1.txt", mode='w')
print(f.write("Welcome from google colab"))
f.close()
```

# append => 'a'

- If file exist then data will be appended.
- if not then new file will be created and data will be written in that file.

```python
f = open("/content/data2.txt", mode='a')
print(f.write("\n Hi from Mohit"))
f.close()
```

# exculsive creation operation => 'x'
- If file exist then it will give fileexist error.
- if not then new file will be created and data will be written in that file.

```python
f = open("/content/data2.txt", mode='x')
print(f.write("\n Hi from Mohit"))
f.close()
```
